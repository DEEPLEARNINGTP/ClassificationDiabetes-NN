```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Activation Functions
def relu(x):
    assert isinstance(x, np.ndarray), "Input to ReLU must be a numpy array"
    result = np.maximum(0, x)
    assert np.all(result >= 0), "ReLU output must be non-negative"
    return result

def relu_derivative(x):
    assert isinstance(x, np.ndarray), "Input to ReLU derivative must be a numpy array"
    result = np.where(x > 0, 1, 0)
    assert np.all((result == 0) | (result == 1)), "ReLU derivative must be 0 or 1"
    return result

def sigmoid(x):
    assert isinstance(x, np.ndarray), "Input to sigmoid must be a numpy array"
    x = np.clip(x, -500, 500)
    result = 1 / (1 + np.exp(-x))
    assert np.all((result >= 0) & (result <= 1)), "Sigmoid output must be in [0,1]"
    return result

def sigmoid_derivative(x):
    assert isinstance(x, np.ndarray), "Input to sigmoid derivative must be a numpy array"
    sig = sigmoid(x)
    result = sig * (1 - sig)
    assert np.all((result >= 0) & (result <= 0.25)), "Sigmoid derivative must be in [0,0.25]"
    return result

# NeuralNetwork Class
class NeuralNetwork:
    def __init__(self, layer_sizes, learning_rate=0.01):
        assert isinstance(layer_sizes, list) and len(layer_sizes) >= 2, "layer_sizes must be a list with at least 2 elements"
        assert all(isinstance(size, int) and size > 0 for size in layer_sizes), "All layer sizes must be positive integers"
        assert isinstance(learning_rate, (int, float)) and learning_rate > 0, "Learning rate must be a positive number"
        self.layer_sizes = layer_sizes
        self.learning_rate = learning_rate
        self.weights = []
        self.biases = []
        np.random.seed(42)
        for i in range(len(layer_sizes) - 1):
            w = np.random.randn(layer_sizes[i], layer_sizes[i+1]) * 0.01
            b = np.zeros((1, layer_sizes[i+1]))
            assert w.shape == (layer_sizes[i], layer_sizes[i+1]), f"Weight matrix {i+1} has incorrect shape"
            assert b.shape == (1, layer_sizes[i+1]), f"Bias vector {i+1} has incorrect shape"
            self.weights.append(w)
            self.biases.append(b)

    def forward(self, X):
        assert isinstance(X, np.ndarray), "Input X must be a numpy array"
        assert X.shape[1] == self.layer_sizes[0], f"Input dimension ({X.shape[1]}) must match input layer size ({self.layer_sizes[0]})"
        self.activations = [X]
        self.z_values = []
        for i in range(len(self.weights) - 1):
            z = np.dot(self.activations[-1], self.weights[i]) + self.biases[i]
            assert z.shape == (X.shape[0], self.layer_sizes[i+1]), f"Z^[i+1] has incorrect shape"
            a = relu(z)
            self.z_values.append(z)
            self.activations.append(a)
        z = np.dot(self.activations[-1], self.weights[-1]) + self.biases[-1]
        assert z.shape == (X.shape[0], self.layer_sizes[-1]), "Output Z has incorrect shape"
        self.z_values.append(z)
        output = sigmoid(z)
        assert output.shape == (X.shape[0], self.layer_sizes[-1]), "Output A has incorrect shape"
        self.activations.append(output)
        return self.activations[-1]

    def compute_loss(self, y_true, y_pred):
        assert isinstance(y_true, np.ndarray) and isinstance(y_pred, np.ndarray), "Inputs to loss must be numpy arrays"
        assert y_true.shape == y_pred.shape, "y_true and y_pred must have the same shape"
        assert np.all((y_true == 0) | (y_true == 1)), "y_true must contain only 0s and 1s"
        m = y_true.shape[0]
        epsilon = 1e-15
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        loss = - (1/m) * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
        assert not np.isnan(loss), "Loss computation resulted in NaN"
        return loss

    def compute_accuracy(self, y_true, y_pred):
        assert isinstance(y_true, np.ndarray) and isinstance(y_pred, np.ndarray), "Inputs to accuracy must be numpy arrays"
        assert y_true.shape == y_pred.shape, "y_true and y_pred must have the same shape"
        predictions = (y_pred >= 0.5).astype(int)
        accuracy = np.mean(predictions == y_true)
        assert 0 <= accuracy <= 1, "Accuracy must be between 0 and 1"
        return accuracy

    def backward(self, X, y, outputs):
        assert isinstance(X, np.ndarray) and isinstance(y, np.ndarray) and isinstance(outputs, np.ndarray), "Inputs to backward must be numpy arrays"
        assert X.shape[1] == self.layer_sizes[0], f"Input dimension ({X.shape[1]}) must match input layer size ({self.layer_sizes[0]})"
        assert y.shape == outputs.shape, "y and outputs must have the same shape"
        m = X.shape[0]
        self.d_weights = [np.zeros_like(w) for w in self.weights]
        self.d_biases = [np.zeros_like(b) for b in self.biases]
        dZ = outputs - y
        assert dZ.shape == outputs.shape, "dZ for output layer has incorrect shape"
        self.d_weights[-1] = (self.activations[-2].T @ dZ) / m
        self.d_biases[-1] = np.sum(dZ, axis=0, keepdims=True) / m
        lambda_reg = 0.01
        for i in range(len(self.weights) - 2, -1, -1):
            dA = dZ @ self.weights[i+1].T
            dZ = dA * relu_derivative(self.z_values[i])
            self.d_weights[i] = (self.activations[i].T @ dZ) / m + (lambda_reg / m) * self.weights[i]
            self.d_biases[i] = np.sum(dZ, axis=0, keepdims=True) / m
        for i in range(len(self.weights)):
            self.weights[i] -= self.learning_rate * self.d_weights[i]
            self.biases[i] -= self.learning_rate * self.d_biases[i]

    def train(self, X, y, X_val, y_val, epochs, batch_size):
        assert isinstance(X, np.ndarray) and isinstance(y, np.ndarray), "X and y must be numpy arrays"
        assert isinstance(X_val, np.ndarray) and isinstance(y_val, np.ndarray), "X_val and y_val must be numpy arrays"
        assert X.shape[1] == self.layer_sizes[0], f"Input dimension ({X.shape[1]}) must match input layer size ({self.layer_sizes[0]})"
        assert y.shape[1] == self.layer_sizes[-1], f"Output dimension ({y.shape[1]}) must match output layer size ({self.layer_sizes[-1]})"
        assert X_val.shape[1] == self.layer_sizes[0], f"Validation input dimension ({X_val.shape[1]}) must match input layer size ({self.layer_sizes[0]})"
        assert y_val.shape[1] == self.layer_sizes[-1], f"Validation output dimension ({y_val.shape[1]}) must match output layer size ({self.layer_sizes[-1]})"
        assert isinstance(epochs, int) and epochs > 0, "Epochs must be a positive integer"
        assert isinstance(batch_size, int) and batch_size > 0, "Batch size must be a positive integer"
        train_losses = []
        val_losses = []
        train_accuracies = []
        val_accuracies = []
        for epoch in range(epochs):
            indices = np.random.permutation(X.shape[0])
            X_shuffled = X[indices]
            y_shuffled = y[indices]
            epoch_loss = 0
            for i in range(0, X.shape[0], batch_size):
                X_batch = X_shuffled[i:i + batch_size]
                y_batch = y_shuffled[i:i + batch_size]
                outputs = self.forward(X_batch)
                epoch_loss += self.compute_loss(y_batch, outputs)
                self.backward(X_batch, y_batch, outputs)
            train_loss = epoch_loss / (X.shape[0] // batch_size)
            train_pred = self.forward(X)
            val_pred = self.forward(X_val)
            train_accuracy = self.compute_accuracy(y, train_pred)
            val_accuracy = self.compute_accuracy(y_val, val_pred)
            val_loss = self.compute_loss(y_val, val_pred)
            train_losses.append(train_loss)
            val_losses.append(val_loss)
            train_accuracies.append(train_accuracy)
            val_accuracies.append(val_accuracy)
            if epoch % 10 == 0:
                print(f"Epoch {epoch}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Train Acc: {train_accuracy:.4f}, Val Acc: {val_accuracy:.4f}")
        return train_losses, val_losses, train_accuracies, val_accuracies

    def predict(self, X):
        assert isinstance(X, np.ndarray), "Input X must be a numpy array"
        assert X.shape[1] == self.layer_sizes[0], f"Input dimension ({X.shape[1]}) must match input layer size ({self.layer_sizes[0]})"
        outputs = self.forward(X)
        predictions = (outputs >= 0.5).astype(int)
        return predictions

# Load and preprocess data
data = pd.read_csv('diabetes.csv')
features = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in features:
    data[col] = data[col].replace(0, data[col].median())
X = data.drop('Outcome', axis=1).values
y = data['Outcome'].values.reshape(-1, 1)
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Initialize and train the neural network
nn = NeuralNetwork(layer_sizes=[X_train.shape[1], 16, 8, 1], learning_rate=0.01)
train_losses, val_losses, train_accuracies, val_accuracies = nn.train(X_train, y_train, X_test, y_test, epochs=100, batch_size=32)

# Evaluate
y_pred = nn.predict(X_test)
print("Classification Report:")
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Plot loss and accuracy curves
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(train_losses, label='Train Loss')
ax1.plot(val_losses, label='Validation Loss')
ax1.set_title('Loss Curves')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()
ax2.plot(train_accuracies, label='Train Accuracy')
ax2.plot(val_accuracies, label='Validation Accuracy')
ax2.set_title('Accuracy Curves')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Fonctions d'activation
def relu(x):
    """
    ReLU activation: max(0, x)
    """
    assert isinstance(x, np.ndarray), "Input to ReLU must be a numpy array" # [cite: 4]
    result = np.maximum(0, x)
    assert np.all(result >= 0), "ReLU output must be non-negative" # [cite: 4]
    return result

def relu_derivative(x):
    """
    Derivative of ReLU: 1 if x > 0, else 0
    """
    assert isinstance(x, np.ndarray), "Input to ReLU derivative must be a numpy array" # [cite: 4]
    result = np.where(x > 0, 1, 0)
    assert np.all((result == 0) | (result == 1)), "ReLU derivative must be 0 or 1" # [cite: 4]
    return result

def sigmoid(x):
    """
    Sigmoid activation: 1 / (1 + exp(-x))
    """
    assert isinstance(x, np.ndarray), "Input to sigmoid must be a numpy array" # [cite: 4]
    result = 1 / (1 + np.exp(-x))
    assert np.all((result >= 0) & (result <= 1)), "Sigmoid output must be in [0, 1]" # [cite: 4]
    return result

def sigmoid_derivative(x): # [cite: 4]
    """
    Derivative of sigmoid: sigmoid(x) * (1 - sigmoid(x))
    """
    assert isinstance(x, np.ndarray), "Input to sigmoid derivative must be a numpy array" # [cite: 4]
    s = sigmoid(x)
    result = s * (1 - s)
    assert np.all((result >= 0) & (result <= 0.25)), "Sigmoid derivative must be in [0, 0.25]" # [cite: 5]
    return result

# Classe Neural Network
class NeuralNetwork: # [cite: 5]
    def __init__(self, layer_sizes, learning_rate=0.01): # [cite: 5]
        """
        Initialize the neural network with given layer sizes and learning rate.
        layer_sizes: List of integers [input_size, hidden1_size, ..., output_size]
        """
        assert isinstance(layer_sizes, list) and len(layer_sizes) >= 2, "layer_sizes must be a list with at least 2 elements" # [cite: 5]
        assert all(isinstance(size, int) and size > 0 for size in layer_sizes), "All layer sizes must be positive integers" # [cite: 5]
        assert isinstance(learning_rate, (int, float)) and learning_rate > 0, "Learning rate must be a positive number" # [cite: 5]

        self.layer_sizes = layer_sizes # [cite: 5]
        self.learning_rate = learning_rate # [cite: 5]
        self.weights = [] # [cite: 5]
        self.biases = [] # [cite: 5]

        # Initialisation des poids et biais
        np.random.seed(42) # [cite: 5]
        for i in range(len(layer_sizes) - 1): # [cite: 5]
            # TODO (This is an example initialization, details might need adjustment based on specific scaling or preferred methods)
            w = np.random.randn(layer_sizes[i], layer_sizes[i+1]) * 0.01
            b = np.zeros((1, layer_sizes[i+1]))
            assert w.shape == (layer_sizes[i], layer_sizes[i+1]), f"Weight matrix {i+1} has incorrect shape" # [cite: 5]
            assert b.shape == (1, layer_sizes[i+1]), f"Bias vector {i+1} has incorrect shape" # [cite: 5]
            self.weights.append(w) # [cite: 5]
            self.biases.append(b) # [cite: 5]

    def forward(self, X): # [cite: 5]
        """
        Forward propagation: Z^[l] = A^[l-1]W^[l] + b^[l], A^[l] = g(Z^[l])
        """
        assert isinstance(X, np.ndarray), "Input X must be a numpy array" # [cite: 5]
        assert X.shape[1] == self.layer_sizes[0], f"Input dimension ({X.shape[1]}) must match input layer size ({self.layer_sizes[0]})" # [cite: 5]

        self.activations = [X] # [cite: 5]
        self.z_values = [] # [cite: 6]
        A = X

        # Couches cachées
        for i in range(len(self.weights) - 1): # [cite: 6]
            Z = np.dot(A, self.weights[i]) + self.biases[i] # [cite: 2]
            assert Z.shape == (X.shape[0], self.layer_sizes[i+1]), f"Z^[{i+1}] has incorrect shape" # [cite: 6]
            self.z_values.append(Z) # [cite: 6]
            A = relu(Z) # [cite: 2]
            self.activations.append(A) # [cite: 6] # TODO (activation is appended)

        # Couche de sortie
        Z_output = np.dot(A, self.weights[-1]) + self.biases[-1] # [cite: 2]
        assert Z_output.shape == (X.shape[0], self.layer_sizes[-1]), "Output Z has incorrect shape" # [cite: 6]
        self.z_values.append(Z_output) # [cite: 6]
        output = sigmoid(Z_output) # [cite: 2]
        assert output.shape == (X.shape[0], self.layer_sizes[-1]), "Output A has incorrect shape" # [cite: 6]
        self.activations.append(output) # [cite: 6]

        return self.activations[-1] # [cite: 6]

    def compute_loss(self, y_true, y_pred): # [cite: 6]
        """
        Binary Cross-Entropy: J = -1/m * sum(y*log(y_pred) + (1-y)*log(1-y_pred))
        """
        assert isinstance(y_true, np.ndarray) and isinstance(y_pred, np.ndarray), "Inputs to loss must be numpy arrays" # [cite: 6]
        assert y_true.shape == y_pred.shape, "y_true and y_pred must have the same shape" # [cite: 6]
        assert np.all((y_true == 0) | (y_true == 1)), "y_true must contain only 0s and 1s" # [cite: 6]
        m = y_true.shape[0]
        # TODO (clip y_pred to avoid log(0))
        epsilon = 1e-15 # Small value to prevent log(0)
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        loss = -1/m * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)) # [cite: 2]
        assert not np.isnan(loss), "Loss computation resulted in NaN" # [cite: 6]
        return loss # [cite: 6]

    def compute_accuracy(self, y_true, y_pred_proba): # [cite: 6]
        """
        Compute accuracy: proportion of correct predictions
        """
        assert isinstance(y_true, np.ndarray) and isinstance(y_pred_proba, np.ndarray), "Inputs to accuracy must be numpy arrays" # [cite: 6]
        assert y_true.shape == y_pred_proba.shape, "y_true and y_pred must have the same shape" # [cite: 6]
        # TODO
        y_pred = (y_pred_proba > 0.5).astype(int) # Convert probabilities to binary predictions
        accuracy = np.mean(y_pred == y_true)
        assert 0 <= accuracy <= 1, "Accuracy must be between 0 and 1" # [cite: 6]
        return accuracy # [cite: 6]

    def backward(self, X, y, outputs): # [cite: 7]
        """
        Backpropagation: compute dW^[l], db^[l] for each layer
        """
        assert isinstance(X, np.ndarray) and isinstance(y, np.ndarray) and isinstance(outputs, np.ndarray), "Inputs to backward must be numpy arrays" # [cite: 7]
        assert X.shape[1] == self.layer_sizes[0], f"Input dimension ({X.shape[1]}) must match input layer size ({self.layer_sizes[0]})" # [cite: 7]
        assert y.shape == outputs.shape, "y and outputs must have the same shape" # [cite: 7]

        m = X.shape[0] # [cite: 7]
        self.d_weights = [np.zeros_like(w) for w in self.weights] # [cite: 7]
        self.d_biases = [np.zeros_like(b) for b in self.biases] # [cite: 7]

        # Couche de sortie
        dZ = outputs - y # [cite: 2]
        assert dZ.shape == outputs.shape, "dZ for output layer has incorrect shape" # [cite: 7]
        self.d_weights[-1] = (self.activations[-2].T @ dZ) / m # [cite: 2]
        self.d_biases[-1] = np.sum(dZ, axis=0, keepdims=True) / m # [cite: 2]

        # Couches cachées (l = L-1, ..., 1)
        for i in range(len(self.weights) - 2, -1, -1): # [cite: 7]
            # TODO
            # dZ^[l] = (dZ^[l+1] (W^[l+1])^T) * ReLU'(Z^[l]) [cite: 2]
            dA_prev = dZ @ self.weights[i+1].T
            dZ = dA_prev * relu_derivative(self.z_values[i]) # [cite: 2]
            self.d_weights[i] = (self.activations[i].T @ dZ) / m # [cite: 2]
            self.d_biases[i] = np.sum(dZ, axis=0, keepdims=True) / m # [cite: 2]

        # TODO: Ajouter une régularisation L2 aux gradients des poids [cite: 7]
        # dW^[l]} + lambda W^[l]} / m, lambda est le coefficient de régularisation [cite: 7]
        # This loop is a placeholder for L2 regularization gradient update
        # lambda_reg = 0.01 # example
        # for i in range(len(self.weights)):
        #     self.d_weights[i] += (lambda_reg / m) * self.weights[i] # [cite: 7]
        pass # [cite: 7]

    def train(self, X, y, X_val, y_val, epochs, batch_size): # [cite: 7]
        """
        Train the neural network using mini-batch SGD, with validation
        """
        assert isinstance(X, np.ndarray) and isinstance(y, np.ndarray), "X and y must be numpy arrays" # [cite: 7]
        assert isinstance(X_val, np.ndarray) and isinstance(y_val, np.ndarray), "X_val and y_val must be numpy arrays" # [cite: 7]
        assert X.shape[1] == self.layer_sizes[0], f"Input dimension ({X.shape[1]}) must match input layer size ({self.layer_sizes[0]})" # [cite: 7]
        assert y.shape[1] == self.layer_sizes[-1], f"Output dimension ({y.shape[1]}) must match output layer size ({self.layer_sizes[-1]})" # [cite: 7]
        assert X_val.shape[1] == self.layer_sizes[0], f"Validation input dimension ({X_val.shape[1]}) must match input layer size ({self.layer_sizes[0]})" # [cite: 8]
        assert y_val.shape[1] == self.layer_sizes[-1], f"Validation output dimension ({y_val.shape[1]}) must match output layer size ({self.layer_sizes[-1]})" # [cite: 8]
        assert isinstance(epochs, int) and epochs > 0, "Epochs must be a positive integer" # [cite: 8]
        assert isinstance(batch_size, int) and batch_size > 0, "Batch size must be a positive integer" # [cite: 8]

        train_losses = [] # [cite: 8]
        val_losses = [] # [cite: 8]
        train_accuracies = [] # [cite: 8]
        val_accuracies = [] # [cite: 8]

        for epoch in range(epochs): # [cite: 8]
            indices = np.random.permutation(X.shape[0]) # [cite: 8]
            X_shuffled = X[indices]
            y_shuffled = y[indices]
            # TODO (shuffling is done) [cite: 8]

            # epoch_loss = 0 # [cite: 8] # Replaced by epoch_train_loss for clarity

            for i in range(0, X.shape[0], batch_size): # [cite: 8]
                # TODO
                X_batch = X_shuffled[i:i+batch_size]
                y_batch = y_shuffled[i:i+batch_size]

                outputs = self.forward(X_batch) # [cite: 8]
                # epoch_loss += self.compute_loss(y_batch, outputs) # [cite: 8] # Corrected to accumulate properly
                self.backward(X_batch, y_batch, outputs) # [cite: 8]

                # Update weights and biases (Mise à jour des poids) [cite: 3]
                for j in range(len(self.weights)):
                    self.weights[j] -= self.learning_rate * self.d_weights[j] # [cite: 3]
                    self.biases[j] -= self.learning_rate * self.d_biases[j] # [cite: 3]

            # Calculer les pertes et accuracies pour l'entraînement et la validation [cite: 8]
            # TODO
            train_outputs = self.forward(X)
            train_loss = self.compute_loss(y, train_outputs)
            train_accuracy = self.compute_accuracy(y, train_outputs)
            train_losses.append(train_loss) # [cite: 8]
            train_accuracies.append(train_accuracy) # [cite: 8]

            val_outputs = self.forward(X_val)
            val_loss = self.compute_loss(y_val, val_outputs)
            val_accuracy = self.compute_accuracy(y_val, val_outputs)
            val_losses.append(val_loss) # [cite: 8]
            val_accuracies.append(val_accuracy) # [cite: 8]


            if epoch % 10 == 0: # [cite: 8]
                print(f"Epoch {epoch}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, " \
                      f"Train Acc: {train_accuracy:.4f}, Val Acc: {val_accuracy:.4f}") # [cite: 8]

        return train_losses, val_losses, train_accuracies, val_accuracies # [cite: 8]

    def predict(self, X): # [cite: 8]
        """
        Predict class labels (0 or 1)
        """
        assert isinstance(X, np.ndarray), "Input X must be a numpy array" # [cite: 9]
        assert X.shape[1] == self.layer_sizes[0], f"Input dimension ({X.shape[1]}) must match input layer size ({self.layer_sizes[0]})" # [cite: 9]
        # TODO
        probabilities = self.forward(X)
        predictions = (probabilities > 0.5).astype(int)
        assert predictions.shape == (X.shape[0], self.layer_sizes[-1]), "Predictions have incorrect shape" # [cite: 9]
        return predictions # [cite: 9]

# Charger et préparer les données
data = pd.read_csv('diabetes.csv') # [cite: 3, 9]
# TODO (Remplacer les zéros non valides par la médiane) [cite: 3, 9]
cols_to_replace_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI'] # [cite: 1, 2]
for col in cols_to_replace_zeros:
    median_val = data[col].median()
    data[col] = data[col].replace(0, median_val)

X_df = data.drop('Outcome', axis=1)
y_df = data['Outcome']

X = X_df.values
y = y_df.values.reshape(-1, 1) # Reshape y to be a column vector

assert X.shape[0] == y.shape[0], "Number of samples in X and y must match" # [cite: 9]
assert X.shape[1] == 8, "Expected 8 features in input data" # [cite: 9]

# Standardisation [cite: 3]
# X_std = (X - mu) / sigma
# TODO
mu = np.mean(X, axis=0) # [cite: 3]
sigma = np.std(X, axis=0) # [cite: 3]
X_scaled = (X - mu) / sigma # [cite: 3]
X = X_scaled # Use scaled X

# Diviser les données en entraînement, validation et test [cite: 3]
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42) # [cite: 3, 9]
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42) # 0.25 * 0.8 = 0.2 # [cite: 9]

assert X_train.shape[0] + X_val.shape[0] + X_test.shape[0] == X.shape[0], "Train-val-test split sizes must sum to total samples" # [cite: 9]

# Créer et entraîner le modèle
layer_sizes = [X_train.shape[1], 16, 8, 1] # [cite: 3, 9]
nn = NeuralNetwork(layer_sizes, learning_rate=0.01) # [cite: 9]
train_losses, val_losses, train_accuracies, val_accuracies = nn.train(X_train, y_train, X_val, y_val, epochs=100, batch_size=32) # [cite: 3, 9]

# TODO: Ajouter une validation croisée pour évaluer la robustesse du modèle [cite: 9]
# TODO: Implémenter l'optimiseur Adam pour une meilleure convergence [cite: 9]

# Predictions et évaluation
y_pred_test = nn.predict(X_test) # [cite: 9]
print("\nRapport de classification (Test set) :") # [cite: 9]
print(classification_report(y_test, y_pred_test)) # [cite: 3, 9]

# Matrice de confusion
cm = confusion_matrix(y_test, y_pred_test) # [cite: 3, 9]
# TODO (Plotting the confusion matrix) [cite: 9]
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non Diabétique', 'Diabétique'], yticklabels=['Non Diabétique', 'Diabétique'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Matrice de Confusion (Test Set)')
plt.show() # [cite: 10]

# Courbes de perte et d'accuracy
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5)) # [cite: 10]

# Courbe de perte [cite: 10]
# TO DO [cite: 10]
ax1.plot(train_losses, label='Train Loss')
ax1.plot(val_losses, label='Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.set_title('Courbe de Perte')
ax1.legend()

# Courbe d'accuracy (implied for ax2)
# TO DO (for ax2, plot accuracies)
ax2.plot(train_accuracies, label='Train Accuracy')
ax2.plot(val_accuracies, label='Validation Accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.set_title('Courbe d\'Accuracy')
ax2.legend()

plt.tight_layout()
plt.show() # [cite: 10]

# Exemple avec 3 couches cachées [cite: 11]
# layer_sizes_ex = [X_train.shape[1], 32, 16, 8, 1] # [cite: 11]
# nn_ex = NeuralNetwork(layer_sizes_ex, learning_rate=0.01) # [cite: 11]
# # The original document's train call for this example is missing validation data.
# # losses_ex = nn_ex.train(X_train, y_train, epochs=100, batch_size=32) # [cite: 11]
# # Assuming a similar call to the main training if validation is desired for this example too:
# # train_losses_ex, val_losses_ex, train_acc_ex, val_acc_ex = nn_ex.train(X_train, y_train, X_val, y_val, epochs=100, batch_size=32)
# print("\nExample with 3 hidden layers demonstration finished if uncommented and adapted.")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Fonctions d'activation
def relu(x):
    """ReLU activation: max(0, x)"""
    return np.maximum(0, x)

def relu_derivative(x):
    """Derivative of ReLU: 1 if x > 0, else 0"""
    return np.where(x > 0, 1, 0)

def sigmoid(x):
    """Sigmoid activation: 1 / (1 + exp(-x))"""
    return 1 / (1 + np.exp(-x))

# Classe Neural Network
class NeuralNetwork:
    def __init__(self, layer_sizes, learning_rate=0.01):
        """
        Initialize the neural network with given layer sizes and learning rate.
        layer_sizes: List of integers [input_size, hidden1_size, ..., output_size]
        """
        self.layer_sizes = layer_sizes
        self.learning_rate = learning_rate
        self.weights = []
        self.biases = []

        # Initialisation des poids et biais
        np.random.seed(42)
        for i in range(len(layer_sizes) - 1):
            w = np.random.randn(layer_sizes[i], layer_sizes[i+1]) * 0.01
            b = np.zeros((1, layer_sizes[i+1]))
            self.weights.append(w)
            self.biases.append(b)

    def forward(self, X):
        """Forward propagation"""
        self.activations = [X]
        self.z_values = []
        A = X

        # Couches cachées
        for i in range(len(self.weights) - 1):
            Z = np.dot(A, self.weights[i]) + self.biases[i]
            self.z_values.append(Z)
            A = relu(Z)
            self.activations.append(A)

        # Couche de sortie
        Z_output = np.dot(A, self.weights[-1]) + self.biases[-1]
        self.z_values.append(Z_output)
        output = sigmoid(Z_output)
        self.activations.append(output)

        return output

    def compute_loss(self, y_true, y_pred, lambda_reg=0.01):
        """
        Binary Cross-Entropy + L2 Regularization:
        J = -1/m * sum(y*log(y_pred) + (1-y)*log(1-y_pred)) + (lambda/2m) * sum(||W||²)
        """
        m = y_true.shape[0]
        epsilon = 1e-15
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        loss = -1/m * np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
        
        # L2 Regularization
        l2_penalty = 0
        for w in self.weights:
            l2_penalty += np.sum(np.square(w))
        
        loss += (lambda_reg / (2 * m)) * l2_penalty
        return loss

    def compute_accuracy(self, y_true, y_pred_proba):
        """Compute accuracy"""
        y_pred = (y_pred_proba > 0.5).astype(int)
        return np.mean(y_pred == y_true)

    def backward(self, X, y, outputs, lambda_reg=0.01):
        """Backpropagation with L2 regularization"""
        m = X.shape[0]
        self.d_weights = [np.zeros_like(w) for w in self.weights]
        self.d_biases = [np.zeros_like(b) for b in self.biases]

        # Couche de sortie
        dZ = outputs - y
        self.d_weights[-1] = (self.activations[-2].T @ dZ) / m + (lambda_reg / m) * self.weights[-1]
        self.d_biases[-1] = np.sum(dZ, axis=0, keepdims=True) / m

        # Couches cachées
        for i in range(len(self.weights) - 2, -1, -1):
            dA_prev = dZ @ self.weights[i+1].T
            dZ = dA_prev * relu_derivative(self.z_values[i])
            self.d_weights[i] = (self.activations[i].T @ dZ) / m + (lambda_reg / m) * self.weights[i]
            self.d_biases[i] = np.sum(dZ, axis=0, keepdims=True) / m

    def train(self, X, y, X_val, y_val, epochs, batch_size, lambda_reg=0.01):
        """Train with mini-batch SGD and L2 regularization"""
        train_losses = []
        val_losses = []
        train_accuracies = []
        val_accuracies = []

        for epoch in range(epochs):
            # Mini-batch SGD
            indices = np.random.permutation(X.shape[0])
            X_shuffled = X[indices]
            y_shuffled = y[indices]

            for i in range(0, X.shape[0], batch_size):
                X_batch = X_shuffled[i:i+batch_size]
                y_batch = y_shuffled[i:i+batch_size]

                outputs = self.forward(X_batch)
                self.backward(X_batch, y_batch, outputs, lambda_reg)

                # Update parameters
                for j in range(len(self.weights)):
                    self.weights[j] -= self.learning_rate * self.d_weights[j]
                    self.biases[j] -= self.learning_rate * self.d_biases[j]

            # Compute metrics
            train_outputs = self.forward(X)
            train_loss = self.compute_loss(y, train_outputs, lambda_reg)
            train_accuracy = self.compute_accuracy(y, train_outputs)
            
            val_outputs = self.forward(X_val)
            val_loss = self.compute_loss(y_val, val_outputs, lambda_reg)
            val_accuracy = self.compute_accuracy(y_val, val_outputs)

            train_losses.append(train_loss)
            val_losses.append(val_loss)
            train_accuracies.append(train_accuracy)
            val_accuracies.append(val_accuracy)

            if epoch % 10 == 0:
                print(f"Epoch {epoch}: Train Loss = {train_loss:.4f}, Val Loss = {val_loss:.4f}, "
                      f"Train Acc = {train_accuracy:.4f}, Val Acc = {val_accuracy:.4f}")

        return train_losses, val_losses, train_accuracies, val_accuracies

    def predict(self, X):
        """Predict class labels (0 or 1)"""
        probabilities = self.forward(X)
        return (probabilities > 0.5).astype(int)

# Chargement et préparation des données
data = pd.read_csv('diabetes.csv')
cols_to_replace_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in cols_to_replace_zeros:
    median_val = data[col].median()
    data[col] = data[col].replace(0, median_val)

X = data.drop('Outcome', axis=1).values
y = data['Outcome'].values.reshape(-1, 1)

# Standardisation
mu = np.mean(X, axis=0)
sigma = np.std(X, axis=0)
X = (X - mu) / sigma

# Split des données
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, stratify=y_temp, random_state=42)

# Entraînement du modèle
layer_sizes = [X_train.shape[1], 16, 8, 1]
nn = NeuralNetwork(layer_sizes, learning_rate=0.01)
train_losses, val_losses, train_accuracies, val_accuracies = nn.train(
    X_train, y_train, X_val, y_val, epochs=100, batch_size=32, lambda_reg=0.01)

# Évaluation
y_pred_test = nn.predict(X_test)
print("\nClassification Report (Test Set):")
print(classification_report(y_test, y_pred_test))

# Matrice de confusion
cm = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Non Diabétique', 'Diabétique'], 
            yticklabels=['Non Diabétique', 'Diabétique'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Matrice de Confusion (Test Set)')
plt.show()

# Courbes d'apprentissage
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(train_losses, label='Train Loss')
ax1.plot(val_losses, label='Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.legend()

ax2.plot(train_accuracies, label='Train Accuracy')
ax2.plot(val_accuracies, label='Validation Accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.legend()

plt.tight_layout()
plt.show()
```
